# PC summary

### What is Predictive coding?


- Predictive coding (PC) is a framework for understanding how the brain adapts to its environment by continuously generating predictions about sensory inputs and correcting itself when errors occur.

- From the theoretical viewpoint of predictive processing, the brain can be seen as a hierarchical model.
- The levels of the hierarchy are implemented by neurons (or clusters of neurons).
- If these levels are compared to regions of the brain:
    - Neurons at one level (or region) attempt to predict the state of neurons at another level (or region).

    - They then adjust their local synaptic parameters based on the difference between their prediction and the observed signal.

- Neurons also use various mechanisms to laterally stimulate and suppress each other.

- These lateral interactions support contextual processing, such as:
    - Grouping visual components of objects in a scene.

    - Segmenting visual components in a scene

### What is Back Propagation?


- During a process of forward propagation, the input data traverses through the different neural layers, where the weight, bias and activation functions transform the data until an output is produced.

- The output may not be correct, so the network needs to caluclate the mistake inorder to learn from it, this is where backpropagation comes in. It is a process of training our model by applying the principle of error correction.

- After forward propagation, the difference between the network output and actual output is calculated. (Loss function) , this error is distributed back through the network providing each neuron in the network a measure of it's contribution to the total error.

- Using this method the back propagation adjust the weights and biases to minimize the error.

### Back Propagation vs Predictive Coding

| Aspect                      | **Backpropagation (BP)**                                         | **Predictive Coding (PC)**                                                         |
| --------------------------- | ---------------------------------------------------------------- | ---------------------------------------------------------------------------------- |
| **Error Signal**            | Uses a **global error** from the output layer, sent backward     | Each layer computes its **own local prediction error**                             |
| **Learning Rule**           | **Non-local**: weight updates depend on distant output error     | **Local**: weights updated using only nearby activity and errors                   |
| **Weight Symmetry**         | Requires **exact symmetry** between forward and backward weights | Does **not require symmetry** between forward and backward pathways                |
| **Function Requirement**    | Needs **differentiable functions** for gradient calculation      | Works with **non-differentiable, flexible dynamics**                               |
| **Connections**             | Purely **feedforward**, no lateral connections                   | Includes **lateral connections** (neurons can influence neighbors)                 |
| **Architecture**            | Rigid, layered structure                                         | Supports **feedback and recurrent loops**, closer to brain circuits                |
| **Biological Plausibility** | Less biologically realistic                                      | More consistent with **brain-like learning** (context-sensitive, energy-efficient) |

### PC Architecture

The 4 Core Steps of Predictive Coding:

- **Generate Top-Down Predictions**: higher levels of the network's hierarchy making predictions about the activity of the levels directly below them. These predictions are not wild guesses; they are based on the model's current understanding of the world, what it has learned from past experiences.This top-down signal essentially represents the brain's "best guess" of what the incoming sensory information should be.

- **Calculate Prediction Error**: The lower level of the hierarchy compares the top-down prediction it receives with its own actual activity. The discrepancy between the prediction and the reality is calculated as a "prediction error". If the prediction is perfect, the error is zero. Any mismatch, however, generates an error signal.

- **Update Neural Activity (Inference)**: The calculated prediction error is then sent up the hierarchy as a bottom-up signal.This error signal drives a rapid update of the neural activity (the "representations" or "causes") at the higher levels. The goal of this step, often called inference, is to adjust the neural activity to create a better prediction that will reduce the error signal. This process happens on a fast timescale, allowing the model to quickly settle on the most likely cause of the sensory input.

- **Update Model Weights (Learning)**: Once the neural activity has been updated to minimize the immediate prediction error, a slower process of learning takes place. The persistent prediction errors are used to adjust the connection strengths (weights) within the network. This is analogous to synaptic plasticity in the brain. By modifying these weights, the network updates its internal generative model, enabling it to make more accurate predictions in the future and thereby reduce future prediction errors.


Core Network Architecture

- The fundamendal Architecture is hierarchical generative model.
- **Generative Hierarchy** : the network is composed of stacked layers of latent variables. Each layer attempts to predict the state directly below it, creating a top-down flow of predictions, starting from abstract higher level representations, and moving towards the concrete lower level sensory data.
- Top-Down Elements: For each layer l, the architecture includes:
    - Weights (W): A weight matrix that connects layer l+1 to layer l. This is used to generate the prediction.
    - Pre-activations (a): Calculated by multiplying the activations of the layer above (x) with the corresponding weight matrix (W).
    - Predictions (x̂): The pre-activation is passed through a non-linear activation function (like ReLU in the paper's implementation) to produce the final prediction for layer l.

- **Prediction Errors** (ε): At each layer, a prediction error is calculated as the simple difference between the actual value of that layer's variables (x) and the top-down prediction (x̂).
- **Loss Function ("Energy")**: The network's overall goal is to minimize the total squared prediction error across all layers. This total error is referred to as "energy."

### What is Hebbian Learning

- Hebbian learning, often summarized as "neurons that fire together, wire together," describes a fundamental principle of synaptic plasticity where the strength of a connection between two neurons increases when they are repeatedly and consistently active at the same time.

- In the context of predictive coding, the modification of synaptic weights (the learning part) is often implemented using a local, Hebbian-style learning rule. Instead of needing a global error signal propagated from a distant output layer, the weight updates are based on the correlation between the activity of pre-synaptic and post-synaptic neurons (the "error" and "prediction" neurons) at a local level.This alignment with Hebbian principles is a fundamental feature that distinguishes predictive coding as a brain-inspired learning algorithm.
